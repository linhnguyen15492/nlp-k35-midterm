```python
import requests
from pathlib import Path

# Thư mục lưu ảnh
save_dir = Path("nlvnpf-0141")
save_dir.mkdir(exist_ok=True)

base_url = (
    "https://lib.nomfoundation.org/site_media/nom/"
    "nlvnpf-0141/large/nlvnpf-0141-{:03d}.jpg"
)


print(save_dir)
for page in range(32, 72):  # 32 -> 71
    url = base_url.format(page)
    filename = save_dir / f"nlvnpf-0141-{page:03d}.jpg"

    try:
        print(f"Downloading {url}")
        r = requests.get(url, timeout=30)
        r.raise_for_status()

        with open(filename, "wb") as f:
            f.write(r.content)

        print(f"✓ Saved: {filename}")

    except Exception as e:
        print(f"✗ Failed: {url}")
        print(e)
```

```python
from pathlib import Path
folder = Path("annamchiluoc")
txt_files = sorted(folder.glob("*.txt"))
output_file = "merged.txt"

with open(folder / output_file, "w", encoding="utf-8") as fout:
    for file in txt_files:
        print(f"Đang nối: {file.name}")
        with open(file, "r", encoding="utf-8") as fin:
            fout.write(fin.read())
            fout.write("\n\n")

print(f"Đã nối {len(txt_files)} file -> {folder / output_file}")
```
